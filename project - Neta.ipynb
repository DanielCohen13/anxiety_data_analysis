```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# קריאת הקובץ
df = pd.read_csv("enhanced_anxiety.csv")  # שנה את שם הקובץ בהתאם לצורך

# קידוד עמודות בינאריות (Yes/No)
binary_mappings = {"Yes": 1, "No": 0}
binary_columns = [
    "Smoking", "Family History of Anxiety", "Dizziness",
    "Medication", "Recent Major Life Event"
]
for col in binary_columns:
    df[col] = df[col].map(binary_mappings)

# קידוד כל שאר העמודות מסוג object לקטגוריות מספריות
for col in df.select_dtypes(include='object').columns:
    df[col] = df[col].astype("category").cat.codes

# חישוב מטריצת המתאמים
correlation_matrix = df.corr()

# ציור ההיטמאפ עם שמות העמודות המקוריים
plt.figure(figsize=(18, 14))
sns.heatmap(
    correlation_matrix,
    annot=True,
    fmt=".2f",
    cmap="coolwarm",
    linewidths=0.5,
    xticklabels=df.columns,
    yticklabels=df.columns
)
plt.title("Heatmap with Full Column Names")
plt.tight_layout()
plt.show()
```

```python
# נניח שעמודת הגיל היא בשם 'Age'
df['Age Group'] = pd.cut(df['Age'], bins=[0, 20, 30, 40, 100], labels=['<20', '20-30', '30-40', '40+'])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid")

# יצירת גריד עם 4 גרפים
fig, axes = plt.subplots(1, 4, figsize=(24, 6), sharex=False)

# גרף 1 – שעות שינה לפי חרדה
sns.barplot(
    data=df,
    x='Anxiety Level (1-10)',
    y='Sleep Hours',
    hue='Anxiety Level (1-10)',
    ax=axes[0],
    palette='Blues',
    dodge=False,
    legend=False
)
axes[0].set_title('Sleep Hours by Anxiety')
axes[0].set_xlabel('Anxiety Level')
axes[0].set_ylabel('Avg Sleep Hours')

# גרף 2 – קפאין לפי חרדה
sns.barplot(
    data=df,
    x='Anxiety Level (1-10)',
    y='Caffeine Intake (mg/day)',
    hue='Anxiety Level (1-10)',
    ax=axes[1],
    palette='Greens',
    dodge=False,
    legend=False
)
axes[1].set_title('Caffeine Intake by Anxiety')
axes[1].set_xlabel('Anxiety Level')
axes[1].set_ylabel('Avg Caffeine (mg)')

# גרף 3 – סטרס לפי חרדה
sns.barplot(
    data=df,
    x='Anxiety Level (1-10)',
    y='Stress Level (1-10)',
    hue='Anxiety Level (1-10)',
    ax=axes[2],
    palette='Reds',
    dodge=False,
    legend=False
)
axes[2].set_title('Stress Level by Anxiety')
axes[2].set_xlabel('Anxiety Level')
axes[2].set_ylabel('Avg Stress Level')

# גרף 4 – טיפולים פסיכולוגיים לפי חרדה
sns.barplot(
    data=df,
    x='Anxiety Level (1-10)',
    y='Therapy Sessions (per month)',
    hue='Anxiety Level (1-10)',
    ax=axes[3],
    palette='Purples',
    dodge=False,
    legend=False
)
axes[3].set_title('Therapy Sessions by Anxiety')
axes[3].set_xlabel('Anxiety Level')
axes[3].set_ylabel('Avg Therapy Sessions')

# התאמת פריסה
plt.tight_layout()
plt.show()

```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid")


# יצירת גריד עם 4 גרפים
fig, axes = plt.subplots(1, 4, figsize=(26, 6), sharex=True)

# גרף 1 – שעות שינה לפי חרדה, מחולק לפי גיל
sns.barplot(
    data=df,
    x='Anxiety Level (1-10)',
    y='Sleep Hours',
    hue='Age Group',
    ax=axes[0],
    palette='Blues'
)
axes[0].set_title('Sleep Hours by Anxiety & Age Group')
axes[0].set_xlabel('Anxiety Level')
axes[0].set_ylabel('Avg Sleep Hours')

# גרף 2 – קפאין לפי חרדה, מחולק לפי גיל
sns.barplot(
    data=df,
    x='Anxiety Level (1-10)',
    y='Caffeine Intake (mg/day)',
    hue='Age Group',
    ax=axes[1],
    palette='Greens'
)
axes[1].set_title('Caffeine Intake by Anxiety & Age Group')
axes[1].set_xlabel('Anxiety Level')
axes[1].set_ylabel('Avg Caffeine (mg)')

# גרף 3 – סטרס לפי חרדה, מחולק לפי גיל
sns.barplot(
    data=df,
    x='Anxiety Level (1-10)',
    y='Stress Level (1-10)',
    hue='Age Group',
    ax=axes[2],
    palette='Reds'
)
axes[2].set_title('Stress Level by Anxiety & Age Group')
axes[2].set_xlabel('Anxiety Level')
axes[2].set_ylabel('Avg Stress Level')

# גרף 4 – טיפולים לפי חרדה, מחולק לפי גיל
sns.barplot(
    data=df,
    x='Anxiety Level (1-10)',
    y='Therapy Sessions (per month)',
    hue='Age Group',
    ax=axes[3],
    palette='Purples'
)
axes[3].set_title('Therapy Sessions by Anxiety & Age Group')
axes[3].set_xlabel('Anxiety Level')
axes[3].set_ylabel('Avg Therapy Sessions')

# התאמת פריסה
plt.tight_layout()
plt.show()
```
